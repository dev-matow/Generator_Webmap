```python
from webmap_generator import generate_webmap

POINTS_URL = "https://drive.google.com/file/d/1XgulX2I2PR9XOiE7_zdYyLAKo6MS0pmX/view?usp=sharing"

def gdrive(url):
    return f"https://drive.google.com/uc?export=download&id={url.split('/d/')[1].split('/')[0]}"

data_list = [
    {"KML_URL": "https://drive.google.com/file/d/1AwYXo3CyvQUmLYLA72avf2DxB1uT1wAa/view?usp=sharing", "site_name": "Demo_Satellite_Only", "ns_id": "NS003"},
    {"KML_URL": "https://drive.google.com/file/d/1AnotherID/view?usp=sharing", "site_name": "Demo_Map_2", "ns_id": "NS004"},
]

results = []

for item in data_list:
    kml_url = gdrive(item["KML_URL"])
    html_file = generate_webmap(
        kml_url=kml_url,
        points_url=gdrive(POINTS_URL),  # ใช้ POINTS_URL เดียวกันทุกครั้ง
        site_name=item["site_name"],
        ns_id=item["ns_id"]
    )
    results.append(html_file)

results
```
